```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
import time
```

```python
archive = 'http://primenjena.psihologija.ff.uns.ac.rs/index.php/pp/issue/view'
issues =[]
paper_urls =[]

for i in range(163,175): 
    url_of_issue = archive + '/' + str(i)
    issues.append(url_of_issue)

for url in issues: 
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, "html.parser")
    for a in soup.select('.obj_article_summary  .title a'):
        link=(a['href'])
        paper_urls.append(link)
    time.sleep(2)
    
paper_title = []
paper_keywords = []
paper_authors = []
paper_afiliations = []
paper_issue = []
paper_date = []

for link in paper_urls: 
    page = requests.get(link)
    content = page.text
    soup = BeautifulSoup(content, "html.parser")
    for title in soup.select(".obj_article_details .page_title"):
        paper_title.append(title.get_text(strip=True))
    for keywords in soup.select(".obj_article_details .item.keywords .value"):
        paper_keywords.append(keywords.get_text(strip=True))
    for authors in soup.select(".obj_article_details .name"):
        paper_authors.append(authors.get_text(strip=True))
    for afiliations in soup.select(".obj_article_details .affiliation"):
        paper_afiliations.append(afiliations.get_text(strip=True))
    for issue in soup.select(".obj_article_details .sub_item"):
        paper_issue.append(issue.get_text(strip=True))
    for date in soup.select(".obj_article_details .entry_details .title"):
        paper_date.append(date.get_text(strip=True))
    time.sleep(2) 
```
